```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("output.txt",sep='\t',header=None)
```

```python
src = np.array(data[0]) # 原始数据
ideal = np.array(data[1]) # 理想数据
dst = np.array(data[2]) # 处理后的数据
```
